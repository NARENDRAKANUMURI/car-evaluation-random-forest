# Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['buying']=le.fit_transform(df.buying)
```

```python
df.head()
```

```python
df.buying.value_counts()
```

```python
df['maint']=le.fit_transform(df.maint)
df['lug_boot']=le.fit_transform(df.lug_boot)
df['safety']=le.fit_transform(df.safety)
```

```python
df.head()
```

```python
df.maint.value_counts()
```

```python
df['outcome']=le.fit_transform(df.outcome)
```

```python
df.head()
```

```python
df.outcome.value_counts()
```

```python
df.doors.value_counts()
```

```python
df.persons.value_counts()
```

```python
## target predictor separation 
x=df.iloc[:,:-1]
y=df.outcome
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
## model creation
from sklearn.ensemble import RandomForestClassifier       ## this is our new model.
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
## model evaluation
from sklearn.metrics import accuracy_score,classification_report
y_pred=model.predict(x_test)
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

```python
# save after training the model
import joblib
joblib.dump(model,"random_forest_model.pkl")
```

```python
import gradio as gr
import joblib
import pandas as pd

# Load the model
model = joblib.load("random_forest_model.pkl")

# Define prediction function
def predict_car_safety(buying, maint, doors, persons, lug_boot, safety):
    input_data = pd.DataFrame([{"buying": buying,
        "maint": maint,
        "doors": doors,
        "persons": persons,
        "lug_boot": lug_boot,
        "safety": safety}])
    prediction = model.predict(input_data)[0]
    return f"Predicted Outcome: {prediction}"

# Define Gradio interface with all 6 inputs
inputs = [
    gr.Dropdown(choices=[0, 1, 2, 3], label="Buying (0: vhigh, 1: high, 2: med, 3: low)"),
    gr.Dropdown(choices=[0, 1, 2, 3], label="Maintenance (0: vhigh, 1: high, 2: med, 3: low)"),
    gr.Dropdown(choices=[0, 1, 2, 3], label="Doors (0: 2, 1: 3, 2: 4, 3: 5more)"),
    gr.Dropdown(choices=[0, 1, 2, 3], label="Persons (0: 2, 1: 4, 2: more, 3: unknown)"),
    gr.Dropdown(choices=[0, 1, 2], label="Lug Boot (0: big, 1: med, 2: small)"),
    gr.Dropdown(choices=[0, 1, 2], label="Safety (0: high, 1: med, 2: low)")
]

# Launch the Gradio interface
gr.Interface(
    fn=predict_car_safety,
    inputs=inputs,
    outputs="text",
    title="Car Evaluation using Random Forest",
    description="Predict car evaluation class based on car features."
).launch()
```
